## Elementi di statistica descrittiva

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Parametri per la generazione delle altezze
num_persone = 1000
altezza_media = 170  # cm
deviazione_standard = 10 # cm

# Genera le altezze usando una distribuzione normale
# np.random.normal(loc=media, scale=dev_std, size=numero_elementi)
altezze = np.random.normal(loc=altezza_media, scale=deviazione_standard, size=num_persone)

# Arrotonda le altezze a due cifre decimali
altezze = np.round(altezze, 2)
```

```python
altezze[:20]
```

```python
print(f"Media: {altezze.mean()}")
print(f"Std: {altezze.std()}")
print(f"Min: {altezze.min()}")
print(f"Max: {altezze.max()}")
print(f"Mediana: {np.median(altezze)}")
print(f"Percentile: {np.percentile(altezze, q=[25, 50, 75])}")

```

```python
fig, ax = plt.subplots(figsize=(12, 6))
ax.boxplot(altezze)
ax.set_title("Altezze")
ax.legend()
plt.tight_layout()
plt.show()
```
