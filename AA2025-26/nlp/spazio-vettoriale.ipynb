## Introduzione al trattamento del testo

```python
import os 
import numpy as np
import nltk 
from nltk.tokenize import word_tokenize
```

```python
folder = "/Users/Flint/Data/recipes/textit"
docids = []
docs = []
for filename in os.listdir(folder):
    if filename.endswith(".txt"):
        with open(os.path.join(folder, filename), 'r') as infile:
            text = infile.read()
            docs.append(text)
            raw = int(filename.replace(".txt", "").split("_")[1])
            docids.append(raw)
```

### Come una macchina legge il testo

```python
def caesar_cipher(text, shift):
    result = ""
    for char in text:
        if char.isalpha():
            # Scegli alfabeto corretto (maiuscolo o minuscolo)
            offset = ord('A') if char.isupper() else ord('a')
            # Applica shift ciclico
            new_char = chr((ord(char) - offset + shift) % 26 + offset)
            result += new_char
        else:
            result += char  # lascia invariati spazi e simboli
    return result
```

### Frequenza

```python
from collections import defaultdict
import pandas as pd
```

```python
def text2chars(text):
    bow = defaultdict(lambda: 0)
    for character in text:
        bow[character] += 1
    return pd.Series(bow) 

def text2words(text):
    bow = defaultdict(lambda: 0)
    for word in word_tokenize(text, language='italian'):
        bow[word] += 1
    return pd.Series(bow) 
```

```python
text1 = "ieri mario è andato a pesca"
text2 = "ieri maria è andata a pesca"
text3 = "oggi gino aveva l'esame di letteratura"
```

```python
bow1 = text2words(text1)
bow2 = text2words(text2)
bow3 = text2words(text3)
```

```python
df = pd.DataFrame([bow1, bow2, bow3]).fillna(0)
df[sorted(df.columns)]
```

```python
print(np.sqrt(np.power(df.loc[0] - df.loc[1], 2).sum()))
print(np.sqrt(np.power(df.loc[0] - df.loc[2], 2).sum()))
```
